```python
# Welcome to your new notebook
# Type here in the cell editor to add code!
df = spark.read.format('delta').load('abfss://878eac13-677f-422e-b9fa-bae10de1d354@onelake.dfs.fabric.microsoft.com/62582fb9-8cbf-4f28-aa83-e002f40371ec/Tables/raw/STORES_GEO_LIST_FINAL')
df.show()
```

```python
from pyspark.sql.functions import split, col

df = df.withColumn("Location_Name", split(col("Location_Name_&_Address"), " - ").getItem(0)) \
       .withColumn("Address", split(col("Location_Name_&_Address"), " - ").getItem(1))
df = df.drop("Location_Name_&_Address")
df.show(truncate=False)
```

```python
df = df.selectExpr("`ST_ID!!` as st_id","`City/Region` as city","`Mgr_Contact#` as mgr_contact","Location_Name as location","Address as address")
df.show()
```
