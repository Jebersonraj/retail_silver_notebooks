```python
# Welcome to your new notebook
# Type here in the cell editor to add code!
df = spark.read.format('delta').load('abfss://878eac13-677f-422e-b9fa-bae10de1d354@onelake.dfs.fabric.microsoft.com/62582fb9-8cbf-4f28-aa83-e002f40371ec/Tables/raw/RTN_TRANS_LOGS')
df.show()
```

```python
df = df.selectExpr("RTN_ID_No as rtn_id_no","`Orgnl_Trxn_ID?` as orgnl_tran_id","Rtn_Date_String as rtn_date","`Rsn_Code_!` as rsn_code")
df.show()
```

```python
from pyspark.sql.functions import to_date, coalesce, col, date_format
df = df.withColumn(
   "rtn_date",
   date_format(
       coalesce(
        to_date(col("rtn_date"),"MM/dd/yyyy"),
        to_date(col("rtn_date"), "dd-MMM-yy"),
        to_date(col("rtn_date"), "MMM dd, yyyy"),
        to_date(col("rtn_date"), "yyyy.MM.dd"), 
        to_date(col("rtn_date"), "dd-MM-yyyy") 
       ),
       "yyyy-MM-dd"
   )
)
df.show()
```
